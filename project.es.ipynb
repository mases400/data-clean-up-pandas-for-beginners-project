#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO

#print(ds.columns)

casa_mas_cara = ds.loc[ds['price'].idxmax()]
direccion = casa_mas_cara['address']
precio = casa_mas_cara['price']
print(f"La casa con direccion en calle {direccion} es la mas cara y su precio es de {precio} USD")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
casa_mas_barata = ds.loc[ds['price'].idxmin()]
direccion = casa_mas_barata['address']
precio = casa_mas_barata['price']
print(f"La casa con direccion en calle {direccion} es la mas barata y su precio es de {precio} USD")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
#print(ds.columns)

casa_mas_superficie = ds.loc[ds["surface"].idxmax()]
direccion_mas_superficie = casa_mas_superficie["address"]
superficie_mas = casa_mas_superficie["surface"]
print(f"La casa mas grande esta ubicada en Calle {direccion_mas_superficie} y su superficie es de {superficie_mas} metros")

casa_menos_superficie = ds.loc[ds["surface"].idxmin()]
direccion_menos_superficie = casa_menos_superficie["address"]
superficie_menos = casa_menos_superficie["surface"]
print(f"La casa mas pequeña esta ubicada en Calle {direccion_menos_superficie} y su superficie es de {superficie_menos} metros")
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

Cuenta el número de poblaciones únicas en la columna 'level5' e imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3, ...`

```python
# TODO

poblaciones_unicas = ds['level5'].dropna().unique()
numero_de_poblaciones = len(poblaciones_unicas)
print(f"El numero de poblaciones unicas es {numero_de_poblaciones}")
print("Poblaciones:")
for poblacion in poblaciones_unicas:
    print(poblacion)
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
filtro = ds.loc[ds['level5'] == 'Arroyomolinos (Madrid)']

media_precios = filtro['price'].mean()

print(f"La media de precios en Arroyomolinos (Madrid) es: {media_precios}")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios

data = filtro["price"]

plt.figure(figsize = (10, 5))

plt.hist(data, bins = 50, alpha = 0.7)

plt.title("Histograma")
plt.show()
```

Lo que puedo aprecia es que el promedio de los precios, ronda los 300,000 USD como nos indico tambien el analisis con codigo, existen algunas propiedades con precios mucho mayores, pero son muy pocos. Tambien se ve que la grafica no es completamente simetrica, los datos muestran que se inclina un poco mas a la derecha hacia los precios por encima de los 300,000
